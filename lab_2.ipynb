```python
import spacy
from spacy.lang.ru.examples import sentences
from spacy import displacy
from pprint import pprint
from collections import defaultdict
from tqdm import tqdm
import pandas as pd
import numpy as np
```

```python
nlp = spacy.load("ru_core_news_sm")
```

```python
newsDF = pd.read_csv('result.CSV')
text_news = newsDF['text']
# text_news = np.array(text_news)
text_news[2]
```

```python
"""
Проходим по каждой строке в списке text_news. Для каждой строчки:

1. Применяем модель обработки естественного языка nlp
2. Извлекаем первое предложение из doc с помощью list(doc.sents)[0].
3. Извлекаем дочерние элементы корневого элемента этого предложения с помощью list(sent0.root.children).
4. Добавляем лемму, текст, зависимость, тег и предка первого дочернего элемента корневого элемента, а также символ "->" в список analisys
5. Увеличиваем count на 1 и повторяем заново

Если есть ошибка обработки, то помещаем None в analisys вместо кортежа

Потом все записываем в файл

"""
```

```python
analisys = []
count = 0
for line in tqdm(text_news):
  try:
    doc = nlp(line)
    sent0 = list(doc.sents)[0]
    childs0 = list(sent0.root.children)
    analisys.append((childs0[0].lemma_, childs0[0].text, childs0[0].dep_, childs0[0].tag_, '->', list(childs0[0].ancestors)[0]))
    count+=1
  except KeyboardInterrupt:
    break
  except:
    analisys.append(None)
    count+=1
    
```

```python
analisys

```

```python
analisys_result = newsDF
analisys_result['syntaxAnalisys'] = analisys
analisys_result

```

```python
analisys_result.to_csv('analisys_result.csv', index=False)
```
